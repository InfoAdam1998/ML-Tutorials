```python
data.split("\t")
```

```python
datalist = list(map(lambda x: x.split("\t"), data.split("\n")))
datalist
```

```python
df = pd.DataFrame(datalist[1:], columns = datalist[0])
df
```

```python
df.info()
```

```python
df.dropna(inplace = True)
df.info()
```

```python
convert_columns = ["level", "page_num", "block_num", "par_num", "line_num", "word_num", "left", "top", "width", "height", "conf"]
df[convert_columns] = df[convert_columns].astype(int)
```

```python
df.dtypes
```

**Draw Bounding Box**

```python
image = img_cv.copy()
level = "paragraph"
for l, x, y, w, h, c, txt in df[["level", "left", "top", "width", "height", "conf", "text"]].values:
  if level == "page":
    if l == 1:
      cv2.rectangle(image, (x,y), (x+w, y+h), (0,0,0), 2)
    else:
        continue

  elif level == "block":
    if l == 2:
      cv2.rectangle(image, (x,y), (x+w, y+h), (255, 0, 0), 2)
    else:
        continue

  elif level == "paragraph":
    if l == 3:
      cv2.rectangle(image, (x,y), (x+w,y+h), (0, 255, 0), 2)
    else:
        continue

  elif level == "line":
    if l == 4:
      cv2.rectangle(image, (x,y), (x+w,y+h), (0,0,255), 2)
    else:
      continue

  elif level == "word":
    if l == 5:
      cv2.rectangle(image, (x,y), (x+w,y+h), (0,0,255), 2)
      cv2.putText(image, txt, (x,y), cv2.FONT_HERSHEY_PLAIN, 1, (255,0,0), 2)
    else:
      continue
```
